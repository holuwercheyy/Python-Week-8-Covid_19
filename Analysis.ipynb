```python
df['daily_new_cases'] = df.groupby('continent')['total_cases'].diff().fillna(0)
```

```python
plt.figure(figsize=(12, 6))

for country in countries_of_interest:
    country_data = df[df['continent'] == country]
    plt.plot(country_data['date'], country_data['daily_new_cases'], label=country)

plt.title('Daily New Cases Comparison')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Add a new column called 'death_rate' to the DataFrame
df['death_rate'] = df['total_deaths'] / df['total_cases']
```

```python
df['death_rate'] = df['total_deaths'] / df['total_cases'].replace(0, pd.NA)
```

```python
df['death_rate'] = df['death_rate'].fillna(0)
```

```python
df_clean = df.dropna(subset=['death_rate'])
```

```python
latest_data = df_clean.sort_values('date').groupby('location').last().reset_index()
```

```python
top_death_rates = latest_data.sort_values('death_rate', ascending=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
countries_of_interest = ['North America','Asia', 'Italy', 'South America']
df_filtered = df[df['continent'].isin(countries_of_interest)]
```

```python
df_filtered = df[(df['location'].isin(countries_of_interest)) & (df['total_vaccinations_per_hundred'].notnull())]
```

```python
latest_vacc = df_filtered.sort_values('date').groupby('location').tail(1)
```

```python
latest_vacc = latest_vacc.dropna(subset=['total_vaccinations_per_hundred'])
```

```python
plt.figure(figsize=(8, 8))

plt.pie(latest_vacc['total_vaccinations_per_hundred'], 
        labels=latest_vacc['location'], 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=['skyblue', 'lightgreen', 'orange', 'pink'])

plt.title('Proportion of % Vaccinated Population (Latest Available Data)')
plt.axis('equal')
plt.show()
```
